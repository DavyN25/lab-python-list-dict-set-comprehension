````python
# 1.Review your code from the previous exercise and identify areas where you can apply comprehension to simplify and streamline your code.
# Define a list of products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Define a function named initialize_inventory that takes products as a parameter. 
#Inside the function, implement the code for initializing the inventory dictionary using a loop and user input
def initialize_inventory(products):
    inventory = {product : int(input(f"Enter the quantity of {product} available: ")) for product in products}
    return inventory
````

````python
inventory = initialize_inventory(products)
````

````python
def get_customer_orders(): 
    customer_orders = set()
    add_product = "yes"
    while add_product.lower() == "yes":
        product_name = input(f"Enter the name of a product to order: ")
        # Add the product name to the "customer_orders" set.
        customer_orders.add(product_name)
        # Ask the user if they want to add another product (yes/no).
        add_product = input(f"Do you want to add another product ? (yes/no) :")
        
    return customer_orders
````

````python
# Define a function named update_inventory that takes customer_orders and inventory as parameters. 
#Inside the function, implement the code for updating the inventory dictionary based on the customer orders.
customer_orders = get_customer_orders()
def update_inventory(customer_orders, inventory):
    updated_inventory= {product : (quantity - 1 if product in  customer_orders else quantity) for product, quantity in inventory.items()}
    return updated_inventory
````

````python
update_inventory(customer_orders, inventory)
````

````python
# Modify the function get_customer_orders so it prompts the user to enter the number of customer orders and gathers the product names using a loop and user input. Use comprehension.
def get_customer_orders():
    # prompts the user to enter the number of customer orders
    number_orders = int (input(f"Enter the number of customer order: "))  
    # Gathers the product names using a List Comprehension 
    orders_list = [input (f"Enter the name of a product that a customer wants to order {i+1} : ") for i in range(number_orders)]  
    return orders_list
````

````python
get_customer_orders()  
````

````python
# 3.Add a new function to calculate the total price of the customer order. For each product in customer_orders, prompt the user to enter the price of that product. 
#Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.
def calculate_total_price (customer_orders):
    total_price = sum(float(input( f"Enter the price of {product} : ")) for product in customer_orders)
    return total_price
````

````python
calculate_total_price (customer_orders)
````

````python
# 4.Modify the update_inventory function to remove the product from the inventory if its quantity becomes zero after fulfilling the customer orders. 
#Use comprehension to filter out the products with a quantity of zero from the inventory.
def update_inventory(customer_orders, inventory):
    updated_inventory= {
        product : 
        (quantity - 1 if product in customer_orders else quantity) 
        for product, quantity in inventory.items()
        if (quantity - 1 if product in customer_orders else quantity)>0
            
}
    return updated_inventory
````

````python
update_inventory(customer_orders, inventory)
````

````python
# We store the result of Calculated the statistics in the variable order_statistics
updated_inventory = update_inventory(customer_orders, inventory)
def print_updated_inventory(updated_inventory):
    print("\nUpdated Inventory:")
    # unpacking  the parameters 
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
````

````python
print_updated_inventory(updated_inventory)
````

````python
# we use the result of statistics made in the previous lab

def calculate_order_statistics(customer_orders, products):
    
    # 1. Total products ordered
    total_product_ordered = len(customer_orders)
    
    # 2. Unique products ordered
    unique_product_ordered = len(set(customer_orders))

    # 3. Total available products
    total_products = len(products)

    # 4. Percentage of unique products ordered
    if total_products >0:
        percentage_unique_p_ordered = (unique_product_ordered / total_products)* 100
    else:
        percentage_unique_p_ordered = 0

    return total_product_ordered, percentage_unique_p_ordered
````

````python
calculate_order_statistics(customer_orders, products)
````

````python
# We store the result of Calculated the statistics in the variable order_statistics
order_statistics = calculate_order_statistics(customer_orders, products)
def print_order_statistics(order_statistics):
    # unpacking  the parameters 
    total_product_ordered, percentage_unique_p_ordered =order_statistics

    print ("order_statistics : ")
    print(f"Total products ordered: {total_product_ordered}")
    print(f"Percentage of unique products ordered: {percentage_unique_p_ordered:.1f}")
````

````python
print_order_statistics(order_statistics)
````
